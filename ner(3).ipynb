```python
#PROGRAM3:   NAMED ENTITY RECOGNITION---
import nltk
from nltk import word_tokenize, pos_tag, ne_chunk, sent_tokenize

input = "The sky woring shades of melted sherbeting as dusk tiptoed in like a shy performer. A dropped pen can start a revolution in a 19 century ."
ner = ne_chunk(pos_tag(word_tokenize(input)))
print(ner)

from nltk.tree import Tree
named_entity = []
for subtree in ner:
    if isinstance(subtree, Tree):
        entity = "".join([token for token, pos in subtree.leaves()])
        named_entity.append(entity)
print(named_entity)
```
